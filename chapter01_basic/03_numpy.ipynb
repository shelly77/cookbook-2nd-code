# 1.3. Introducing the multidimensional array in NumPy for fast array computations

```python
import random
import numpy as np
```

```python
n = 10
x = [random.random() for _ in range(n)]
y = [random.random() for _ in range(n)]
```

```python
x, y
```

```python
z = [x[i] + y[i] for i in range(n)]
z
```

```python
%timeit [x[i] + y[i] for i in range(n)]
```

```python
xa = np.array(x)
ya = np.array(y)
```

```python
xa
```

```python
za = xa + ya
za
```

```python
%timeit xa + ya
```

```python
%timeit sum(x)  # pure Python
```

```python
%timeit np.sum(xa)  # NumPy, so much faster
```

```python
d = [abs(x[i] - y[j])
     for i in range(10)
     for j in range(10)]
```

```python
d
```

```python
da = np.abs(xa[:10, np.newaxis] - ya[:10])
```

```python
da
```

```python
%timeit [abs(x[i] - y[j]) \
         for i in range(10) \
         for j in range(10)]
```

```python
%timeit np.abs(xa[:,np.newaxis] - ya)
```
